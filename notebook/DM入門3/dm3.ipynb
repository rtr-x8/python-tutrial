## データとは

レコードとフィールドからなる多次元データ。  
レコードは属性や変数等フィールドの集合からなる。  
通常はレコード間に依存性はない。

## フィールドの値

数値、カテゴリ、順序、大小関係、バイナリ、テキストなどがある

## データの記述統計
### 中心傾向
平均など。最大値最小値を除外して外れ値を除外する

中央値：外れ値の値を受けにくい  
最頻値：カテゴリカルデータに用いることが多い  
範囲（レンジ）：最大-最小  
四分位：Q1,2,3  
分散：平均からのばらつき  
標準偏差：分散の正の平方根  
標準化：異なるデータを統一的な指標に変換する  

## 標本と母集団
標本は、全体（母集団）からサンプリングした部分的なデータのこと  
標本から母集団を推定する。  

推定時は普遍分散を行う。-1する。n-1は自由度のこと。

## 可視化
度数分布（ヒストグラム）：全体的な分布  
階級数：区分の数、階級幅：区分の幅

```python

import math


def avg_med_var(inp: list)->list:
    inp.sort()
    c = len(inp)
    s = sum(inp)
    mean = s/c
    md = 0
    if c % 2 == 1:
        md = inp[c//2]
    else:
        md = (inp[c//2]+inp[c//2+1])/2
    b = 0
    for i in inp:
        b += math.abs(i-mean)**2
    b /= c
    
    return [mean, md, b]

print(avg_med_var([1,4,3,2,5,6,8,7,9,10]))
print(avg_med_var([1,4,3,2,5,6,8,7,9,10,11]))
```
